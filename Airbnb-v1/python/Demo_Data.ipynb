```python
#add dependencies
import os
import numpy as np
import pandas as pd
import matplotlib as plt
import json
from pprint import pprint
```

```python
#load json files
filepath = os.path.join("..", "data", "mash-api", "120320", "top_nbh_overview-120320j.json")
with open(filepath) as jsonfile:
    neighborhood_overview_json = json.load(jsonfile)
    
pprint(neighborhood_overview_json)
```

```python
neighboorhood_overview_df= pd.DataFrame(neighborhood_overview_json).transpose()
neighboorhood_overview_df.head()
```

```python
datatest = neighboorhood_overview_df.iloc['content']['id']
print(datatest)
```

```python
neighboorhood_data = []
for i in range(len(neighboorhood_overview_df)):
    
        if (neighboorhood_overview_df.iloc[i]['content']['walkscore'] is not None):
            walk_score = int(neighboorhood_overview_df.iloc[i]['content']['walkscore'])
        else:
            walk_score = 0
        
        if (neighboorhood_overview_df.iloc[i]['content']['num_of_airbnb_properties'] is not None):
            airbnb_count = int(neighboorhood_overview_df.iloc[i]['content']['num_of_airbnb_properties'])
        else:
            airbnb_count = 0
            
        if (neighboorhood_overview_df.iloc[i]['content']['num_of_traditional_properties'] is not None):
            other_count = int(neighboorhood_overview_df.iloc[i]['content']['num_of_traditional_properties'])
        else:
            other_count = 0
    
        specific_data = {
            "id":neighboorhood_overview_df.iloc[i]['content']['id'],
            "name":neighboorhood_overview_df.iloc[i]['content']['name'],
            "city":neighboorhood_overview_df.iloc[i]['content']['city'],
            "county":neighboorhood_overview_df.iloc[i]['content']['county'],
            "state":neighboorhood_overview_df.iloc[i]['content']['state'],
            "lat":neighboorhood_overview_df.iloc[i]['content']['latitude'],
            "lon":neighboorhood_overview_df.iloc[i]['content']['longitude'],
            "walkscore":walk_score,
            "airbnb_count":airbnb_count,
            "other_count":other_count,
            "avg_occupancy":neighboorhood_overview_df.iloc[i]['content']['avg_occupancy'],
            "median_price":neighboorhood_overview_df.iloc[i]['content']['median_price'],
            "sqft_price":neighboorhood_overview_df.iloc[i]['content']['price_per_sqft']
        }
        neighboorhood_data.append(specific_data)
        
neighboorhood_df = pd.DataFrame(neighboorhood_data)
neighboorhood_df.index = neighboorhood_df['id']
del neighboorhood_df['id']
neighboorhood_df
```
