```python
# Interactive plots
%matplotlib notebook

# Dependencies and Setup
import requests
import random
import json
import ast
import os
import datetime
import pandas as pd
#import numpy as np
#import matplotlib.pyplot as plt
#import scipy.stats as st
#from citipy import citipy
#from pprint import pprint

# Set by selecting a datasource import type (CSV or API)
source = 'api'
#source = 'csv'

# Remember to update the config file with your API key
from api_keys import mash_api_key

# Set variable for current date using datetime library
today = datetime.date.today()

# Set Mashvisor API URL for rental rates
url = "https://mashvisor-api.p.rapidapi.com/city/neighborhoods"

# Set API variables
state = "TX"
city = "Houston"
source = "airbnb"

# Set todays date as active and formatted as mm/dd/yy for API import
date = today.strftime('%m/%d/%y')
datef = today.strftime('%m%d%y')

# Set a specific date as active parameter for CSV import
#date = '09/12/20'
#datef = '091220'

# Print active date
print(date)
#print(mash_api_key)
```

```python
# Create and call Mashvisor API query string for rental rates
querystring = f'{url}/{state}/{city}'
headers = {
    'x-rapidapi-key': mash_api_key,
    'x-rapidapi-host': "mashvisor-api.p.rapidapi.com"
    }
response = requests.request("GET", querystring, headers=headers)
print(response.text)
```

```python
nhdata = response.json()
print(nhdata)
```

```python
with open(f'../data/mash-api/{datef}/lst_nbh-{city}-{datef}.json', 'w') as f1:
    json.dump(nhdata,f1)
    print("The neighborhood json file is created")
f1.close()
```

```python
f = open(f'../data/mash-api/{datef}/lst_nbh-{city}-{datef}.json', 'r')
s = f.read()
s = s.replace("\'", "\"")
s = s.replace("null", "''")
airstr = json.dumps(s)
airjson = json.loads(airstr)
airbnb = ast.literal_eval(airjson)
df = pd.DataFrame(airbnb)
print(df)
print("")
samples = df.iloc[0,1]
print(samples)
print("")
df_nbh = pd.DataFrame(df.iloc[1,1])
del df_nbh['is_village']
df_nbh.index = df_nbh['id']
del df_nbh['id']
df_nbh
```

```python
result = df_nbh.to_json(orient="index")
parsed = json.loads(result)
nbh_json = json.dumps(parsed, indent=4)
print(nbh_json)
```

```python
with open(f'../data/mash-api/{datef}/lst_nbh-{city}-details-{datef}.json', 'w') as f1:
    json.dump(nbh_json,f1,indent=4)
    print("The nbh json file is created")
f1.close()
```
