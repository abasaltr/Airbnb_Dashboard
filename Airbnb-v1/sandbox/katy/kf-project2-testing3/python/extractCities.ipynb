```python
# Interactive plots
%matplotlib notebook

# Dependencies and Setup
import requests
import random
import json
import ast
import os
import datetime
import pandas as pd
#import numpy as np
#import matplotlib.pyplot as plt
#import scipy.stats as st
#from citipy import citipy
#from pprint import pprint

# Set by selecting a datasource import type (CSV or API)
source = 'api'
#source = 'csv'

# Remember to update the config file with your API key
from api_keys import mash_api_key

# Set variable for current date using datetime library
today = datetime.date.today()

# Set Mashvisor API URL for rental rates
url = "https://mashvisor-api.p.rapidapi.com/trends/cities"

# Set API variables
states = ["TX","CA","FL","NY","DC"]
source = "airbnb"

# Set todays date as active and formatted as mm/dd/yy for API import
date = today.strftime('%m/%d/%y')
datef = today.strftime('%m%d%y')

# Set a specific date as active parameter for CSV import
#date = '09/12/20'
#datef = '091220'

# Print active date
print(date)
#print(mash_api_key)
```

```python
# Create and call Mashvisor API query string for rental rates
stateResponse = []
for state in states:
    querystring = {"state":f'{state}',"page":"1","items":"5"}
    print(f'extracting {state} top cities...')
    headers = {
        'x-rapidapi-key': mash_api_key,
        'x-rapidapi-host': "mashvisor-api.p.rapidapi.com"
    }
    response = requests.request("GET", url, headers=headers, params=querystring)
    stateResponse.append(response.json())
    print('data merged')
    print('')
```

```python
print(stateResponse[0])
```

```python
for i in range(len(states)):
    with open(f'../data/mash-api/{datef}/topAirbnbCities-{states[i]}-{datef}.json', 'w') as f1:
        json.dump(stateResponse[i],f1)
        print(f'The cities json file in {states[i]} is created...')
    f1.close()
```

```python
statesArray = []
for i in range(len(states)):
    f = open(f'../data/mash-api/{datef}/topAirbnbCities-{states[i]}-{datef}.json', 'r')
    s = f.read()
    s = s.replace("\'", "\"")
    s = s.replace("null", "''")
    airstr = json.dumps(s)
    airjson = json.loads(airstr)
    airbnb = ast.literal_eval(airjson)
    statesArray.append(airbnb)

topCities = []
for st in statesArray:
    for i in range(st['content']['total_page_results']):
            citydata = {}
            citydata = {'city': st['content']['cities'][i]['city'],
               'state': st['content']['cities'][i]['state'],
                'occupancy': st['content']['cities'][i]['occupancy'],
                'total_listing': st['content']['cities'][i]['total_listing']
             }
            topCities.append(citydata)
    
df_cities = pd.DataFrame(topCities)
df_cities    
df_cities.index = df_cities['city']
del df_cities['city']
df_cities    
```

```python
result = df_cities.to_json(orient="index")
parsed = json.loads(result)
cities_json = json.dumps(parsed, indent=4)
print(cities_json)
```

```python
with open(f'../data/mash-api/{datef}/topAirbnbCities-{datef}.json', 'w') as f1:
    json.dump(cities_json,f1,indent=4)
    print("The top airbnb cities json file is created")
f1.close()
```
